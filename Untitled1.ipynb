```python
import pandas as pd

ipl_file_path = '/content/IPL2020_Match_data.csv'
ipl_df = pd.read_csv(ipl_file_path)
display(ipl_df.head())
```

```python

team_wins = ipl_df['Winner'].value_counts()


most_winning_team = team_wins.idxmax()
max_wins = team_wins.max()
print(f"The team with the most wins is '{most_winning_team}' with {max_wins} wins.")

display(team_wins)
```

### Bar Chart: Total Wins per Team

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 7))
sns.barplot(x=team_wins.index, y=team_wins.values, hue=team_wins.index, palette='viridis', legend=False)
plt.title('Total Wins per Team (IPL 2020)')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns



# The print statement is no longer needed after identifying the correct column name.
# print(ipl_df.columns)

toss_decisions = ipl_df['Toss_decision'].value_counts()

plt.figure(figsize=(8, 5))
sns.barplot(x=toss_decisions.index, y=toss_decisions.values, hue=toss_decisions.index, palette='coolwarm', legend=False)
plt.title('Distribution of Toss Decisions (IPL 2020)')
plt.xlabel('Toss Decision')
plt.ylabel('Number of Decisions')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
plt.pie(team_wins.values, labels=team_wins.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.viridis(range(len(team_wins))))
plt.title('Proportion of Wins per Team (IPL 2020)')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

#Sample dataset
data={
    'Name':['Amrita','Mohit','Chand','Dev'],
    'Age':[20,19,22,21],
    'Gender':['F','M','M','M'],
    'Department':['HR','IT', 'Finance','IT']
}

df=pd.DataFrame(data)
print(df)
#separate category





# Separate categrical and numerical columns
categorical_columns=df.select_dtypes(include=['object']).columns
numerical_columns=df.select_dtypes(include=['int64','float64']).columns
```

```python
import numpy as np

df['Log_Age']=np.log(df['Age'])
df['Age_zscore']=(df['Age']-df['Age'].mean())/df['Age'].std()
print(df)
```

```python
df['Age_Group']=pd.cut(df['Age'],bins=[20,30,40,50],labels=['20-30','30-40','40-50'])
print(df)

#Standarization (z-score normalization)
df['Age_zscore']=(df['Age']-df['Age'].mean())/df['Age'].std()
print(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

gender_counts = df['Gender'].value_counts()

plt.figure(figsize=(7, 5))
sns.barplot(x=gender_counts.index, y=gender_counts.values, hue=gender_counts.index, palette='pastel', legend=False)
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```
