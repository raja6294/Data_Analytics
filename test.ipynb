```python
import pandas as pd

file_path = '/content/Indian Summers - Over the years-2007-11.csv'
df = pd.read_csv(file_path)
df['Date'] = pd.to_datetime(df['Date'])
summer_months = [4, 5, 6]
summer_df = df[df['Date'].dt.month.isin(summer_months)]

display(summer_df.head())
display(f"Total records for summer months: {len(summer_df)}")
display(f"Years included in summer data: {summer_df['Date'].dt.year.unique().tolist()}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Create a figure and a set of subplots
fig, axes = plt.subplots(1, 3, figsize=(20, 6))

# Plot histogram for 'temp'
sns.histplot(summer_df['temp'], bins=20, kde=True, ax=axes[0], color='skyblue')
axes[0].set_title('Distribution of Average Temperature (Summer)')
axes[0].set_xlabel('Average Temperature (°C)')
axes[0].set_ylabel('Frequency')

# Plot histogram for 'tempmax'
sns.histplot(summer_df['tempmax'], bins=20, kde=True, ax=axes[1], color='salmon')
axes[1].set_title('Distribution of Maximum Temperature (Summer)')
axes[1].set_xlabel('Maximum Temperature (°C)')
axes[1].set_ylabel('Frequency')

# Plot histogram for 'tempmin'
sns.histplot(summer_df['tempmin'], bins=20, kde=True, ax=axes[2], color='lightgreen')
axes[2].set_title('Distribution of Minimum Temperature (Summer)')
axes[2].set_xlabel('Minimum Temperature (°C)')
axes[2].set_ylabel('Frequency')

# Adjust layout to prevent overlapping titles/labels
plt.tight_layout()
plt.show()
```
